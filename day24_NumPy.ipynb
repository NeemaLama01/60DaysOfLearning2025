```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt

img=imageio.imread('/content/doraemon.jpg')
print(img.shape)
print(img.dtype)
plt.imshow(img)
plt.show()
```

```python
import numpy as np
#reading from raw file
img=np.fromfile('/content/doraemon.jpg',dtype=np.uint8)
print(img.shape)
```

```python
#getting min,max,mean values
print(img.max())
print(img.min())
print(img.mean())
```

```python
# height, width, channels
x,y,_=img.shape
# Crop the image: from x//3 to -x//8 vertically, and y//3 to -y//8 horizontally
crop = img[x//3: -x//8, y//3: -y//8]
plt.imshow(crop)
plt.show()
```
