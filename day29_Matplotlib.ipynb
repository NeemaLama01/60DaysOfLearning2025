# Bar Plot in Matplotlib

A bar plot uses rectangular bars to represent data categories withbar length or height proportional to values. The x-axis typically shows the categories while y-axis shows the values associated with those categories.

Syntax:

plt.bar(x,height,width,bottom,align)

```python
#Customizing Bar Colors
import matplotlib.pyplot as plt
import numpy as np

grade=['9','10','11','12']
percentage=[90,85,80,75]

plt.bar(grade,percentage,color='blue')
plt.title('Grade wise Percentage')
plt.xlabel('Grade')
plt.ylabel('Percentage')
plt.show()
```

```python
#Creating Horizontal Bar Plots
import matplotlib.pyplot as plt
import numpy as np

fruits=['Mango','Guava','Apple','Litchi']
Price=[190,85,280,75]

plt.barh(fruits,Price)
plt.title('Fruits sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
#Adjusting Bar width
import matplotlib.pyplot as plt
import numpy as np

grade=['9','10','11','12']
percentage=[90,85,80,75]

plt.bar(grade,percentage,width=0.1,color='green')
plt.title('Grade wise Percentage')
plt.xlabel('Grade')
plt.ylabel('Percentage')
plt.show()
```

#Matplotlib Scatter

Scatter plots are powerful tools for visualizing relationship between two numerical variables, cluster or distribute using **matplotlib.pyplot.scatter()**.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,45,32,7,1,30])
y=np.array([77,88,22,35,42,70])

plt.scatter(x,y)
plt.title('Basic Scatter Plot')
plt.xlabel('X val')
plt.ylabel('Y val')
plt.grid()
plt.show()
```

```python
#using different colors for each group
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,45,32,7,1,30])
y=np.array([77,88,22,35,42,70])

x1=np.array([2,40,52,37,10,31])
y1=np.array([27,18,62,55,46,20])

plt.scatter(x,y,color='black',label='Group1')
plt.scatter(x1,y1,color='orange',label='Group2')
plt.title('Different colors for each group')
plt.xlabel('X val')
plt.ylabel('Y val')
plt.grid()
plt.show()
```

```python
#Scatter Plot with Varying Colors and Sizes
import matplotlib.pyplot as plt
import numpy as np

x=np.array([12,45,32,7,1,30])
y=np.array([77,88,22,35,42,70])

size=[200,90,140,250,110,60]
color=['red','brown','blue','cyan','purple','green']
plt.scatter(x,y,s=size,c=color,edgecolors='w',linewidths=1)
plt.title('Scatter Plot with Varying Colors and Sizes')
plt.xlabel('X val')
plt.ylabel('Y val')
plt.grid()
plt.show()
```
