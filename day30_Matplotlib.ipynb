# Matplotlib Histogram
It represents data provided in the form of some groups/ numerical data distribution where X-axis refers to the bin ranges while Y-axis refers to frequency.

Syntax: matplotlib.pyplot.hist()


```python
#Basic Histogram in Matplotlib
import matplotlib.pyplot as plt
import numpy as np

#generate random data
data=np.random.randn(1000)

#plot histogram
plt.hist(data, bins=20,color='pink',edgecolor='brown')

#add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

#Display the plot
plt.show()
```

```python
#Customized Histogram in Matplotlib with Density Plot
import seaborn as sns
import numpy as np

#generate random data
data=np.random.randn(1000)

#plot histogram
sns.histplot(data, bins=20,kde=True,color='orange',edgecolor='brown')

#add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

#Display the plot
plt.show()
```

# Pie Chart using Matplotlib

It is a circular statistical plot to display only one series of data where each slice is proportional to the quantity it represents.

syntax: matplotlib.pyplot.pie()

```python
#plotting a pie chart in matplotlib
import matplotlib.pyplot as plt
import numpy as np

#creating dataset
students=['Nima','Rabi','Rohan','Ngima']
data=[34,55,80,65]

#creating a plot
plt.pie(data,labels=students,shadow=True)

#show plot
plt.show()
```
