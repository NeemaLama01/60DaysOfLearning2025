```python
#Pandas Access Dataframe
import pandas as pd
#creating data using python dictionary
data={'Name':['Nima','Rabi','Anita','Raunak'],
      'Age':[21,22,23,34],
      'Gender':['F','M','F','M']}
df=pd.DataFrame(data)
df
```

```python
#accessing column from DataFrame
age_column=df['Age']
age_column
```

```python
#accessing multiple columns
data=df[['Age','Gender']]
data
```

```python
#accessing rows by Index(iloc for positional indexing)
second_row=df.iloc[1]
second_row
```

```python
#accessing multiple rows
subset=df.loc[0:2,['Name','Gender']]
print(subset)
```

```python
#accessing rows based on conditions
filtered_data=df[df['Age']>30]
filtered_data
```

```python
#accessing specific cells
index_2=df.at[0,'Name']
index_2
```

```python
#query() using boolean expression
result=df.query("Age > 20 and Gender =='F'")
result
```
