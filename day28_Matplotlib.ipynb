# Introduction to Matplotlib

It is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats.

```python
import matplotlib.pyplot as plt

x=[0,1,2,3,4]
y=[2,4,6,8,10]

plt.plot(x,y)
plt.show()
```

```python
x=[0,2,4,6,8]
y=[0,4,16,36,64]

fig,ax=plt.subplots()
ax.plot(x,y,marker='o', label='Data Points')

ax.set_title("Basic components of Matplotlib Figure")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4])
y=x*2

plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Title")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

# first plot with X and Y data
plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```
