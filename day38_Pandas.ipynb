#Grouping and Aggregating with Pandas

When working with large datasets it's used to group and summarize the data to make analysis easier.

#Aggregation in Pandas
Aggregation means applying a mathematical function to summarize data. Using agg().

```python
import pandas as pd

df = pd.DataFrame([[9, 4, 8, 9], [8, 10, 7, 6],[7, 6, 8, 5]],columns=['Maths',  'English', 'Science', 'History'])
print(df)
```

```python
print(df.sum()) #Summing Up All Values (sum())
print(df.describe()) #Getting a Summary (describe()), provides all important statistics in one go.
```

```python
df.agg(['sum', 'min', 'max']) #Applying Multiple Aggregations at Once (agg())
```
