#3D plotting using Matplotlib

Visualizing data involving three variables oftern requires three-dimensional plotting.

Python's Matplotlib library through its mpl_toolkits.mplot3d toolkit, provides powerful support for 3D visualizations.

```python
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')
plt.show()
```

**3D line plot**

It connects points in three-dimesional space tovisualize a continuous path useful to shw how a vraiable evolves over time or space in 3D.

```python
#3D line plot
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')

z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'blue')
ax.set_title('3D Line Plot')
plt.show()
```

**3D Scatter plot**

It displays individual data points in three dimensions, helpful for spotting trends or clusters.

```python
#3D Scatter plot
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x + y  # Color array based on x and y

ax.scatter(x, y, z, c=c)
ax.set_title('3D Scatter Plot')
plt.show()
```
