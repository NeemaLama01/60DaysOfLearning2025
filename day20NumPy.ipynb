# Binomial Distribution

It is a fundamental concept in probability and statistics which models the number of successes in a fixed number of independent trials where each trial has only two possible outcomes success or failure.

Syntax- numpy.random.binomial(n,p,size=None)


```python
#single random number

import numpy as np
random_num=np.random.binomial(n=10,p=0.5) #n means num of trials, p means probability of success
print(random_num)
```

```python
# array/multiple random numbers

random_nums=np.random.binomial(n=10,p=0.5,size=5)
print(random_nums)
```
