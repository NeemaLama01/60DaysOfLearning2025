# Uniform Distribution in NumPy

It is used when all numbers in a range have the same chance of being picked.

Syntax: numpy.random.uniform(low=0.0, high=1.0,size=None)

```python
# single random number

import numpy as np
random_num=np.random.uniform()
print(random_num)
```

```python
# multiple/array of random numbers
random_nums=np.random.uniform(size=5)
print(random_nums)
```

```python
#visualizing uniform distribution
import matplotlib.pyplot as plt
import seaborn as sns
low=10
high=20
size=1000
data=np.random.uniform(low=low,high=high,size=size)
sns.histplot(data,bins=30,color='pink',edgecolor='skyblue')
plt.title(f"Uniform Distribution(Range: {low} to {high})")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

The above image shows a Unifrom Distribution between 10 and 20. this means every number in that range is equally likely to happen. The bars in the histogram show that values ranging from 10 to 20 appear about the same number of times.

# Exponential Distribution in NumPy

It is a fundamental concept in probability and statistics which describes the time events in a Poisson process where events occur continuously and independently at a constant average rate.

Syntax: numpy.random.exponential(scale=1.0,size=None)

```python
#single random number
random_number=np.random.exponential()
print(random_number)
```

```python
#generate multiple random numbers
random_nums=np.random.exponential(size=5)
print(random_nums)
```

```python
#Visualizing the exponential distribution
scale=2
size=1000
data=np.random.exponential(scale=scale,size=size)
sns.histplot(data,bins=30,kde=True,color='orange',edgecolor='red')
plt.title(f"Exponential distribution (Scale={scale})")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

The above image shows an Exponential Distribution with a sacle of parameter of 2. The orange bars shows how frequently each values appeared in the data while smooth orange curve is a density curve to visualize the shape of distribution.
