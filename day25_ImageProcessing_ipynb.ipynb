**Flip Image**

using flipud() function of the numpy module image can be flipped in up-down direction.


```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
#reading image
image=imageio.imread('/content/doraemon.jpg')
#flipping image
flip=np.flipud(image)
plt.imshow(flip)
plt.show()
```

**Rotating Images**

Using ndimage.rotate() function from scipy can be used to rotate image at specific angle.

```python
from scipy import ndimage
#rotating the image
rotate=ndimage.rotate(image,60)
plt.imshow(rotate)
plt.show()
```

**Blurring Image**

It is a process of reducing noise level in the image using gaussian or unicorn filter.

```python
#blurring the image
blur=ndimage.gaussian_filter(image,sigma=7) #sigma is parameter controlling blur
plt.imshow(blur)
plt.show()
```
